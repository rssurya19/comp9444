```python
class RSICDDataset(Dataset):
    def __init__(self, image_ids, image_captions, image_id_to_filename, images_path,
                 tokenizer, max_length=50, transform=None):
        self.image_ids = image_ids
        self.image_captions = image_captions
        self.image_id_to_filename = image_id_to_filename
        self.images_path = images_path
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.transform = transform

        # Create flattened dataset (one entry per image-caption pair)
        self.data = []
        for img_id in image_ids:
            filename = image_id_to_filename[img_id]
            for caption in image_captions[img_id]:
                self.data.append((img_id, filename, caption))

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_id, filename, caption = self.data[idx]

        # Load image
        img_path = os.path.join(self.images_path, filename)
        try:
            image = Image.open(img_path).convert('RGB')
            if self.transform:
                image = self.transform(image)
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
            # Return a black image as fallback
            image = torch.zeros(3, 224, 224)

        # Additional error handling for image loading
        if image.numel() == 0:
            print(f"Warning: Empty image tensor for {filename}")
            image = torch.zeros(3, 224, 224)

        # Tokenize caption
        caption_tokens = self.tokenizer.encode(
            caption,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_tensors='pt'
        ).squeeze()

        return {
            'image': image,
            'caption': caption_tokens,
            'caption_text': caption,
            'image_id': img_id
        }

# Image transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

# Initialize tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
```

```python
class ImageCaptioningModel(nn.Module):
    def __init__(self, vocab_size, hidden_size=768, num_layers=12):
        super(ImageCaptioningModel, self).__init__()

        # Image encoder (ResNet-50)
        self.image_encoder = resnet50(pretrained=True)
        self.image_encoder.fc = nn.Linear(self.image_encoder.fc.in_features, hidden_size)

        # Language model (GPT-2)
        config = GPT2Config(
            vocab_size=vocab_size,
            n_embd=hidden_size,
            n_layer=num_layers,
            n_head=12,
            n_positions=1024
        )
        self.language_model = GPT2LMHeadModel(config)

        # Projection layer to combine image and text features
        self.image_projection = nn.Linear(hidden_size, hidden_size)
        self.dropout = nn.Dropout(0.1)

    def forward(self, images, captions=None):
        batch_size = images.size(0)

        # Encode images
        image_features = self.image_encoder(images)  # [batch_size, hidden_size]
        image_features = self.image_projection(image_features)
        image_features = self.dropout(image_features)

        if captions is not None:
            # Training mode
            # Expand image features to match sequence length
            seq_len = captions.size(1)
            image_features = image_features.unsqueeze(1).repeat(1, seq_len, 1)

            # Get text embeddings
            inputs_embeds = self.language_model.transformer.wte(captions)

            # Combine image and text features
            combined_embeds = inputs_embeds + image_features

            # Forward through language model
            outputs = self.language_model(inputs_embeds=combined_embeds, labels=captions)
            return outputs
        else:
            # Inference mode
            return image_features

# Initialize model
vocab_size = tokenizer.vocab_size
model = ImageCaptioningModel(vocab_size).to(device)
print(f"Model initialized with {sum(p.numel() for p in model.parameters())} parameters")
```

```python
# Install and setup evaluation metrics
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer

# For CIDER and SPICE (more complex setup)
class EvaluationMetrics:
    def __init__(self):
        self.smoothing = SmoothingFunction()
        self.rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'],
                                                   use_stemmer=True)

    def compute_bleu(self, reference, hypothesis):
        """Compute BLEU scores (1-4)"""
        reference_tokens = [reference.split()]
        hypothesis_tokens = hypothesis.split()

        bleu_scores = {}
        for i in range(1, 5):
            weights = [1.0/i] * i + [0.0] * (4-i)
            bleu_scores[f'bleu_{i}'] = sentence_bleu(
                reference_tokens, hypothesis_tokens,
                weights=weights, smoothing_function=self.smoothing.method1
            )

        return bleu_scores

    def compute_meteor(self, reference, hypothesis):
        """Compute METEOR score"""
        try:
            # Clean the text before computing METEOR
            ref_clean = reference.strip().lower()
            hyp_clean = hypothesis.strip().lower()

            # Skip if either is empty
            if not ref_clean or not hyp_clean:
                return 0.0

            score = meteor_score([ref_clean], hyp_clean)
            return score if not np.isnan(score) else 0.0
        except Exception as e:
            # Debug print for METEOR issues
            if hasattr(self, '_meteor_error_count'):
                self._meteor_error_count += 1
            else:
                self._meteor_error_count = 1
                print(f"METEOR computation error (will only show once): {e}")
            return 0.0

    def compute_rouge(self, reference, hypothesis):
        """Compute ROUGE scores"""
        scores = self.rouge_scorer.score(reference, hypothesis)
        return {
            'rouge_1': scores['rouge1'].fmeasure,
            'rouge_2': scores['rouge2'].fmeasure,
            'rouge_l': scores['rougeL'].fmeasure
        }

    def compute_all_metrics(self, references, hypotheses):
        """Compute all metrics for a list of reference-hypothesis pairs"""
        all_metrics = defaultdict(list)

        for ref, hyp in zip(references, hypotheses):
            # BLEU scores (1-4)
            bleu_scores = self.compute_bleu(ref, hyp)
            for key, score in bleu_scores.items():
                all_metrics[key].append(score)

            # METEOR score
            meteor = self.compute_meteor(ref, hyp)
            all_metrics['meteor'].append(meteor)

            # ROUGE scores
            rouge_scores = self.compute_rouge(ref, hyp)
            for key, score in rouge_scores.items():
                all_metrics[key].append(score)

        # Average all scores
        avg_metrics = {}
        for key, scores in all_metrics.items():
            avg_metrics[key] = np.mean(scores)

        return avg_metrics

# Initialize evaluator
evaluator = EvaluationMetrics()
```

```python
class EarlyStopping:
    def __init__(self, patience=7, min_delta=0.001, restore_best_weights=True):
        self.patience = patience
        self.min_delta = min_delta
        self.restore_best_weights = restore_best_weights
        self.best_loss = None
        self.counter = 0
        self.best_weights = None

    def __call__(self, val_loss, model):
        if self.best_loss is None:
            self.best_loss = val_loss
            self.save_checkpoint(model)
        elif val_loss < self.best_loss - self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
            self.save_checkpoint(model)
        else:
            self.counter += 1

        if self.counter >= self.patience:
            if self.restore_best_weights:
                model.load_state_dict(self.best_weights)
            return True
        return False

    def save_checkpoint(self, model):
        self.best_weights = model.state_dict().copy()

# Training parameters
BATCH_SIZE = 16
LEARNING_RATE = 5e-5
NUM_EPOCHS = 15
PATIENCE = 3

# Create datasets using the configured paths
train_dataset = RSICDDataset(train_ids, image_captions, image_id_to_filename,
                           IMAGES_DIRECTORY, tokenizer, transform=transform)
val_dataset = RSICDDataset(val_ids, image_captions, image_id_to_filename,
                         IMAGES_DIRECTORY, tokenizer, transform=transform)
test_dataset = RSICDDataset(test_ids, image_captions, image_id_to_filename,
                          IMAGES_DIRECTORY, tokenizer, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

# Optimizer and scheduler
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=0.01)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)

# Early stopping
early_stopping = EarlyStopping(patience=PATIENCE)

print(f"Training batches: {len(train_loader)}")
print(f"Validation batches: {len(val_loader)}")
print(f"Test batches: {len(test_loader)}")
```

```python
def generate_caption(model, image, tokenizer, max_length=50, device='cuda', temperature=0.7):
    """Generate caption for a single image with improved sampling"""
    model.eval()
    with torch.no_grad():
        # Get image features
        image = image.unsqueeze(0).to(device)
        image_features = model(image)

        # Start with BOS token (using EOS as BOS for GPT-2)
        generated = torch.tensor([[tokenizer.eos_token_id]], device=device)

        for step in range(max_length):
            # Get current embeddings
            with torch.no_grad():
                inputs_embeds = model.language_model.transformer.wte(generated)

                # Add image features to first token
                if inputs_embeds.size(1) == 1:
                    inputs_embeds = inputs_embeds + image_features.unsqueeze(1)

                # Forward pass
                outputs = model.language_model(inputs_embeds=inputs_embeds)
                predictions = outputs.logits[0, -1, :]

                # Apply temperature and get next token
                if temperature > 0:
                    predictions = predictions / temperature
                    # Use top-k sampling for better diversity
                    top_k = 50
                    top_k_predictions, top_k_indices = torch.topk(predictions, top_k)
                    probs = torch.softmax(top_k_predictions, dim=-1)
                    next_token_idx = torch.multinomial(probs, 1)
                    next_token = top_k_indices[next_token_idx]
                else:
                    next_token = torch.argmax(predictions, dim=-1, keepdim=True)

                # Stop if end token or padding token
                if next_token.item() in [tokenizer.eos_token_id, tokenizer.pad_token_id]:
                    break

                # Append token
                generated = torch.cat([generated, next_token.unsqueeze(0)], dim=1)

        # Decode caption
        caption = tokenizer.decode(generated[0], skip_special_tokens=True)

        # Clean up the caption
        caption = caption.strip()
        if not caption:
            caption = "a remote sensing image"  # Fallback caption

        return caption

def evaluate_model(model, dataloader, tokenizer, evaluator, device, max_samples=None):
    """Evaluate model on dataset"""
    model.eval()
    references = []
    hypotheses = []

    with torch.no_grad():
        for i, batch in enumerate(tqdm(dataloader, desc="Evaluating")):
            if max_samples and i >= max_samples:
                break

            images = batch['image'].to(device)
            true_captions = batch['caption_text']

            # Generate captions
            for j in range(images.size(0)):
                generated_caption = generate_caption(model, images[j], tokenizer, device=device)
                references.append(true_captions[j])
                hypotheses.append(generated_caption)

    # Compute metrics
    metrics = evaluator.compute_all_metrics(references, hypotheses)
    return metrics, references, hypotheses

def show_sample_generations(model, dataset, tokenizer, device, num_samples=3):
    """Show sample generations during training"""
    model.eval()
    indices = np.random.choice(len(dataset), min(num_samples, len(dataset)), replace=False)

    print("\n" + "="*60)
    print("SAMPLE GENERATIONS:")
    print("="*60)

    for i, idx in enumerate(indices):
        sample = dataset[idx]
        image = sample['image']
        true_caption = sample['caption_text']

        # Generate caption
        generated_caption = generate_caption(model, image, tokenizer, device=device)

        print(f"Sample {i+1}:")
        print(f"  True:      {true_caption}")
        print(f"  Generated: {generated_caption}")
        print()

    print("="*60)
```

```python
# Training history
history = {
    'train_loss': [],
    'val_loss': [],
    'val_metrics': [],
    'learning_rates': []
}

best_val_loss = float('inf')
best_metrics = {}

print("Starting training...")
print("="*50)

for epoch in range(NUM_EPOCHS):
    # Training phase
    model.train()
    train_loss = 0.0
    train_pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{NUM_EPOCHS} [Train]')

    for batch in train_pbar:
        images = batch['image'].to(device)
        captions = batch['caption'].to(device)

        optimizer.zero_grad()

        # Forward pass
        outputs = model(images, captions)
        loss = outputs.loss

        # Backward pass
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()

        train_loss += loss.item()
        train_pbar.set_postfix({'loss': loss.item()})

    avg_train_loss = train_loss / len(train_loader)

    # Validation phase
    model.eval()
    val_loss = 0.0
    val_pbar = tqdm(val_loader, desc=f'Epoch {epoch+1}/{NUM_EPOCHS} [Val]')

    with torch.no_grad():
        for batch in val_pbar:
            images = batch['image'].to(device)
            captions = batch['caption'].to(device)

            outputs = model(images, captions)
            loss = outputs.loss
            val_loss += loss.item()
            val_pbar.set_postfix({'loss': loss.item()})

    avg_val_loss = val_loss / len(val_loader)

    # Evaluate on validation set (subset for speed)
    val_metrics, _, _ = evaluate_model(model, val_loader, tokenizer, evaluator, device, max_samples=50)

    # Update learning rate
    scheduler.step(avg_val_loss)
    current_lr = optimizer.param_groups[0]['lr']

    # Save history
    history['train_loss'].append(avg_train_loss)
    history['val_loss'].append(avg_val_loss)
    history['val_metrics'].append(val_metrics)
    history['learning_rates'].append(current_lr)

    # Print epoch results with all BLEU scores
    print(f"Epoch {epoch+1}/{NUM_EPOCHS}:")
    print(f"  Train Loss: {avg_train_loss:.4f}")
    print(f"  Val Loss: {avg_val_loss:.4f}")
    print(f"  BLEU-1: {val_metrics.get('bleu_1', 0):.4f}")
    print(f"  BLEU-2: {val_metrics.get('bleu_2', 0):.4f}")
    print(f"  BLEU-3: {val_metrics.get('bleu_3', 0):.4f}")
    print(f"  BLEU-4: {val_metrics.get('bleu_4', 0):.4f}")
    print(f"  METEOR: {val_metrics.get('meteor', 0):.4f}")
    print(f"  ROUGE-L: {val_metrics.get('rouge_l', 0):.4f}")
    print(f"  Learning Rate: {current_lr:.2e}")

    # Show sample generations every few epochs
    if (epoch + 1) % 2 == 0 or epoch == 0:
        show_sample_generations(model, val_dataset, tokenizer, device, num_samples=2)

    print("-" * 50)

    # Save best model
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        best_metrics = val_metrics.copy()

        # Save best model
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'val_loss': avg_val_loss,
            'metrics': val_metrics
        }, '/content/best_model/best_model.pth')

        print(f"  ✓ New best model saved (Val Loss: {avg_val_loss:.4f})")

    # Early stopping check
    if early_stopping(avg_val_loss, model):
        print(f"Early stopping triggered after epoch {epoch+1}")
        break

    print()

print("Training completed!")
print(f"Best validation loss: {best_val_loss:.4f}")
print("Best metrics:", best_metrics)
```

```python
# Advanced CIDER metric implementation
from collections import defaultdict, Counter
import math

class CIDErMetric:
    def __init__(self, n=4, sigma=6.0):
        self.n = n
        self.sigma = sigma

    def cook_refs(self, refs):
        """Cook the references."""
        return [self.cook_sent(sent) for sent in refs]

    def cook_sent(self, sent):
        """Cook a sentence."""
        words = sent.split()
        counts = defaultdict(int)
        for k in range(1, self.n + 1):
            for i in range(len(words) - k + 1):
                ngram = tuple(words[i:i + k])
                counts[ngram] += 1
        return counts

    def compute_doc_freq(self, refs_list):
        """Compute document frequencies."""
        document_frequency = defaultdict(float)
        for refs in refs_list:
            for ref in refs:
                for ngram in set(ref.keys()):
                    document_frequency[ngram] += 1
        return document_frequency

    def compute_cider(self, refs_list, hyps_list):
        """Compute CIDEr score."""
        assert len(refs_list) == len(hyps_list)

        document_frequency = self.compute_doc_freq(refs_list)
        scores = []

        for refs, hyp in zip(refs_list, hyps_list):
            vec, norm, length = self.sim(refs, hyp, document_frequency, len(refs_list))
            scores.append(norm)

        return np.mean(scores)

    def sim(self, refs, hyp, document_frequency, ref_len):
        """Compute similarity."""
        def counts2vec(cnts):
            vec = [defaultdict(float) for _ in range(self.n)]
            length = 0
            norm = [0.0 for _ in range(self.n)]
            for (ngram, term_freq) in cnts.items():
                df = np.log(max(1.0, document_frequency[ngram]))
                n = len(ngram) - 1
                vec[n][ngram] = float(term_freq) * (ref_len - df)
                norm[n] += pow(vec[n][ngram], 2)
                if n == 1:
                    length += term_freq
            norm = [np.sqrt(n) for n in norm]
            return vec, norm, length

        def sim_one(vec_hyp, vec_ref, norm_hyp, norm_ref, length_hyp, length_ref):
            delta = float(length_hyp - length_ref)
            val = np.array([0.0 for _ in range(self.n)])
            for n in range(self.n):
                for (ngram, count) in vec_hyp[n].items():
                    val[n] += min(vec_hyp[n][ngram], vec_ref[n][ngram]) * vec_ref[n][ngram]

                if (norm_hyp[n] != 0) and (norm_ref[n] != 0):
                    val[n] /= (norm_hyp[n] * norm_ref[n])

                assert not math.isnan(val[n])
                val[n] *= np.e**(-(delta**2)/(2*self.sigma**2))
            return val

        # Convert to vectors
        vec_hyp, norm_hyp, length_hyp = counts2vec(hyp)

        scores = np.array([0.0 for _ in range(self.n)])
        for ref in refs:
            vec_ref, norm_ref, length_ref = counts2vec(ref)
            scores += sim_one(vec_hyp, vec_ref, norm_hyp, norm_ref, length_hyp, length_ref)

        scores_avg = scores / len(refs)
        mean_score = np.mean(scores_avg)
        return scores_avg, mean_score, length_hyp

# SPICE metric (simplified version)
class SPICEMetric:
    def __init__(self):
        # This is a simplified version - full SPICE requires Java dependencies
        pass

    def compute_spice(self, references, hypotheses):
        """Simplified SPICE score based on semantic similarity"""
        # This is a placeholder - real SPICE requires scene graph parsing
        # For now, we'll use a combination of word overlap and semantic features
        scores = []

        for ref, hyp in zip(references, hypotheses):
            ref_words = set(ref.lower().split())
            hyp_words = set(hyp.lower().split())

            if len(ref_words.union(hyp_words)) == 0:
                score = 0.0
            else:
                score = len(ref_words.intersection(hyp_words)) / len(ref_words.union(hyp_words))

            scores.append(score)

        return np.mean(scores)

# Enhanced evaluation with CIDER and SPICE
class EnhancedEvaluationMetrics(EvaluationMetrics):
    def __init__(self):
        super().__init__()
        self.cider_metric = CIDErMetric()
        self.spice_metric = SPICEMetric()

    def compute_enhanced_metrics(self, references, hypotheses):
        """Compute all metrics including CIDER and SPICE"""
        # Get base metrics
        base_metrics = self.compute_all_metrics(references, hypotheses)

        # Prepare data for CIDER
        refs_list = []
        hyps_list = []

        for ref, hyp in zip(references, hypotheses):
            refs_cooked = self.cider_metric.cook_refs([ref])
            hyp_cooked = self.cider_metric.cook_sent(hyp)
            refs_list.append(refs_cooked)
            hyps_list.append(hyp_cooked)

        # Compute CIDER
        cider_score = self.cider_metric.compute_cider(refs_list, hyps_list)

        # Compute SPICE (simplified)
        spice_score = self.spice_metric.compute_spice(references, hypotheses)

        # Combine all metrics
        enhanced_metrics = base_metrics.copy()
        enhanced_metrics['cider'] = cider_score
        enhanced_metrics['spice'] = spice_score

        return enhanced_metrics

# Update evaluator
enhanced_evaluator = EnhancedEvaluationMetrics()
```
